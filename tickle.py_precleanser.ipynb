```python
#Some code I pieced together to pre cleanse data for my ASMR machine learning project, splits audio into 8 sec, 16 bit, 16kHz chunks as described in the Mehri et al SamplerRNN paper
#code on how to iterate through a directory for generating training samples https://stackoverflow.com/questions/65857983/how-can-i-split-an-audio-file-into-multiple-audio-wav-files-from-folder
from pydub import AudioSegment 
from pydub.utils import make_chunks
import os
import librosa
import numpy
import sox

def process_sudio(file_name):
    
    #downsamples and changes bitrate of audio file
    fileDownsamp = sox.Transformer()
    fileDownsamp.set_output_format(rate=16000, bits = 16) #downsamples audio and crushes bit depth
    #new folder for downsampled files
    ds_file_name = "./downsampled/" + file_name
    fileDownsamp.build(input_filepath = file_name,sample_rate_in = 16000, output_filepath = ds_file_name)
    
    #loads downsampled audio and sets chunk size
    myaudio = AudioSegment.from_file(ds_file_name, "wav") 
    chunk_length_ms = 8000 # pydub calculates in millisec 
    pad_ms = 8000
    chunks = make_chunks(myaudio,chunk_length_ms) #Make chunks of 8 sec 
  
    for i, chunk in enumerate(chunks): 
        
        chunk_name = './chunked/' + file_name + "_{0}.wav".format(i) 
        print ("exporting", chunk_name) 
        chunk.export(chunk_name, format="wav")
        
        #zero padding if statement
        #pad code adapted from https://stackoverflow.com/questions/52841335/how-can-i-pad-wav-file-to-specific-length
        if (librosa.get_duration(filename=(chunk_name)))!=8:
            print("zero padding end of file")
            #this should be the chunk of audio that is currently bieng processed if it is less than 8 secs
            audio = AudioSegment.from_wav(chunk_name)
            silence = AudioSegment.silent(duration=pad_ms-len(audio)+1)
            padded = audio + silence
            padded.export(chunk_name, format='wav') #hopefully this replaces and deletes the old file automatically could throw error                              
all_file_names = os.listdir()
try:
    os.makedirs('chunked')# creating folders named chunked and downsampled
    os.makedirs('downsampled')
except:
    pass
for each_file in all_file_names:
    if ('.wav' in each_file):
        process_sudio(each_file)
```
